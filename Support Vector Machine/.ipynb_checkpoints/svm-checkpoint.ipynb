```python
import pandas as  pd
import numpy as np
import matplotlib.pyplot as plot
```

```python
data = pd.read_csv('heart.csv', index_col = 0)
data.head()
```

```python
x = data[['oldpeak','slope']].values.reshape([-1,2])
y = data['target'].to_numpy()
#y = np.where(y <= 0, -1, 1)
x_train = x[:272]
y_train = y[:272]
x_test = x[272:]
y_test = y[272:]

color = data['target'].apply(lambda x: 'red' if x == 0 else 'green')
data.plot(kind='scatter', x='oldpeak', y='slope', c=color)
plot.show()
np.array(x_train)
np.array(y_train)
print(y_train)
```

```python
def predict(x, w, b):
    classification = np.sign(np.dot(x,w) + b)
    return classification
```

```python
def cost_func(w, b, x, y):
    cf = y *(np.dot(x, w) + b)
    return cf
        
```

```python
def fit(x, y, b):
    w = np.zeros(len(x[0]))
    counter1 = 0
    counter2 = 0
    LR = 0.00000001
    lambda_df = 1 / 1000
    for i in range(1000):
        for j, x_term in enumerate(x):
            cost_f = cost_func(w, b, x_term, y[j])
            #print(cost_f)
            if cost_f >= 1:
                w -= LR * (2 * np.dot(lambda_df, w))
            else:
                w += (np.dot(y[j], x_term) - 2 * np.dot(lambda_df, w))
                     
    return w
```

```python
b = np.random.normal()
w = fit(x_train, y_train, b)
print(w)
Y_pred = predict(x_test, w, b)
print(y_test)
print(Y_pred)
error = 0
for i in range(31):
    if(Y_pred[i] != y_test[i]):
        error += 1
        
print((31-error)/31)
```
